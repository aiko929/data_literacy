```python
# Installation der notwendigen Bibliotheken
!pip install matplotlib seaborn pandas numpy

# Import der Bibliotheken
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Image
```

**Beispieldaten einlesen**

```python
np.random.seed(42)
sample = np.random.normal(70, 10, 100)  # 100 Werte mit Mittelwert 70 und Standardabweichung 10
sample2 = np.random.normal(75, 12, 100)  # Zweites Sample für Vergleich
# Beispielhafte zweite Variable für Scatterplot
time = np.arange(1, 101)  # Simulierte Zeitpunkte
age = np.random.randint(20, 80, 100)  # Zufällige Alterswerte zwischen 20 und 80
```

**Oder die Daten aus einer CSV-Datei laden**

```python
csv_name = 'heart.csv'  # Datei muss in Colab hochgeladen werden oder Pfad anpassen
df = pd.read_csv(csv_name)
sample = df['heartrate']  # Spalte anpassen, falls notwendig
```

**Diagramme erstellen**


---


**Histogramm**

```python
plt.figure(figsize=(8, 5))
plt.hist(sample, bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Werte')
plt.ylabel('Häufigkeit')
plt.title('Histogramm der Daten')
plt.grid(True)
plt.show()
```

```python
# Histogramm mit zwei Datensaetzen
plt.figure(figsize=(8, 5))
plt.hist(sample, bins=15, color='skyblue', edgecolor='black', alpha=0.6, label='Sample 1')
plt.hist(sample2, bins=15, color='orange', edgecolor='black', alpha=0.6, label='Sample 2')
plt.xlabel('Werte')
plt.ylabel('Häufigkeit')
plt.title('Histogramm der Daten (Zwei Datensätze)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Histogramm als Dichteplot
plt.figure(figsize=(8, 5))
sns.kdeplot(sample, fill=True, color='skyblue', alpha=0.6, label='Sample 1')
sns.kdeplot(sample2, fill=True, color='orange', alpha=0.6, label='Sample 2')
plt.xlabel('Werte')
plt.ylabel('Dichte')
plt.title('Dichteplot der Daten (Zwei Datensätze)')
plt.legend()
plt.grid(True)
plt.show()
```

**Boxplot**

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x=sample, color='lightcoral')
plt.xlabel('Werte')
plt.title('Boxplot der Daten')
plt.show()
```

```python
# Boxplot zweier Datensaetze
plt.figure(figsize=(6, 5))
sns.boxplot(data=[sample, sample2], palette=['lightcoral', 'lightblue'])
plt.xticks([0, 1], ['Sample 1', 'Sample 2'])
plt.xlabel('Datensatz')
plt.title('Boxplot der Daten')
plt.show()
```

**Violinplot**

```python
plt.figure(figsize=(6, 5))
sns.violinplot(x=sample, color='lightgreen')
plt.xlabel('Werte')
plt.title('Violinplot der Daten')
plt.show()
```

```python
# Violinplot zweier Datensätze
plt.figure(figsize=(6, 5))
sns.violinplot(data=[sample, sample2], palette=['lightgreen', 'lightblue'])
plt.xticks([0, 1], ['Sample 1', 'Sample 2'])
plt.xlabel('Datensatz')
plt.title('Violinplot der Daten')
plt.show()
```

**Scatterplot**

```python
# Scatterplot (Falls zweite Variable vorhanden ist)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=age, y=sample, color='blue')
plt.xlabel('Alter')
plt.ylabel('Werte')
plt.title('Scatterplot: Alter vs. Werte')
plt.grid(True)
plt.show()

# Unten für die Nutzung einer CSV

#if 'age' in df.columns:
#    plt.figure(figsize=(8, 5))
#    sns.scatterplot(x=df['age'], y=sample, color='blue')
#    plt.xlabel('Alter')
#    plt.ylabel('Herzrate')
#    plt.title('Scatterplot: Alter vs. Herzrate')
#    plt.grid(True)
#    plt.show()
```

```python
# Scatterplot (Alter vs. Sample-Daten, beide Gruppen)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=age, y=sample, color='blue', label='Sample 1')
sns.scatterplot(x=age, y=sample2, color='red', label='Sample 2')
plt.xlabel('Alter')
plt.ylabel('Werte')
plt.title('Scatterplot: Alter vs. Werte (Zwei Datensätze)')
plt.legend()
plt.grid(True)
plt.show()
```

**Lineplot**

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=time, y=sample, marker='o', color='purple')
plt.xlabel('Zeit')
plt.ylabel('Werte')
plt.title('Liniendiagramm der Werte über die Zeit')
plt.grid(True)
plt.show()

# Unten für die Nutzung einer CSV

#if 'time' in df.columns:
#    plt.figure(figsize=(8, 5))
#    sns.lineplot(x=df['time'], y=sample, marker='o', color='purple')
#    plt.xlabel('Zeit')
#    plt.ylabel('Herzrate')
#    plt.title('Liniendiagramm der Herzrate über die Zeit')
#    plt.grid(True)
#    plt.show()
```

```python
# Lineplot (Zwei Datensätze über Zeit)
plt.figure(figsize=(8, 5))
sns.lineplot(x=time, y=sample, marker='o', color='purple', label='Sample 1')
sns.lineplot(x=time, y=sample2, marker='o', color='green', label='Sample 2')
plt.xlabel('Zeit')
plt.ylabel('Werte')
plt.title('Liniendiagramm der Werte über die Zeit (Zwei Datensätze)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Heatmap (Korrelationsmatrix eines simulierten Datensatzes)
data = pd.DataFrame({
    'Alter': age,
    'Werte Sample 1': sample,
    'Werte Sample 2': sample2,
    'Zeit': time
})
plt.figure(figsize=(8, 6))

sns.heatmap(data.corr(), annot=True, cmap="YlGnBu", fmt='.2f')
# Andere Farben sind unten aufgelistet
#sns.heatmap(data.corr(), annot=True, fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap="Blues", fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap="BuPu", fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap="Greens", fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap="Oranges", fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap="Reds", fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap="Purples", fmt='.2f')
#sns.heatmap(data.corr(), annot=True, cmap="YlOrBr", fmt='.2f')


plt.title('Heatmap der Korrelationen')
plt.show()
```

**Balkendiagramm**

```python
# Balkendiagramm für Vergleich zweier Datensätze
categories = ['Sample 1', 'Sample 2']
means = [np.mean(sample), np.mean(sample2)]
plt.figure(figsize=(6, 5))
plt.bar(categories, means, color=['blue', 'orange'])
plt.xlabel('Datensatz')
plt.ylabel('Mittelwert')
plt.title('Balkendiagramm der Mittelwerte')
plt.show()
```

```python
# Gruppiertes Balkendiagramm für verschiedene Kategorien
categories = ['A', 'B', 'C', 'D']
values1 = np.random.randint(10, 50, 4)
values2 = np.random.randint(10, 50, 4)
x = np.arange(len(categories))
width = 0.4

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, values1, width, label='Gruppe 1', color='blue')
plt.bar(x + width/2, values2, width, label='Gruppe 2', color='orange')
plt.xticks(ticks=x, labels=categories)
plt.xlabel('Kategorie')
plt.ylabel('Werte')
plt.title('Gruppiertes Balkendiagramm')
plt.legend()
plt.show()
```

**Kreisdiagramm**

```python
# Kreisdiagramm (Pie Chart) für den Anteil beider Datensätze
sizes = [len(sample)*2, len(sample2)] # Die Groesse von sample 1 ist aus Anschauungszwecken verdoppelt worden
labels = ['Sample 1', 'Sample 2']
colors = ['blue', 'orange']
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Anteile der Stichproben')
plt.show()
```

**Stacked Area**

```python
# Stacked Area Plot (Gestapeltes Flächendiagramm)
plt.figure(figsize=(8, 5))
plt.stackplot(time, sample, sample2, labels=['Sample 1', 'Sample 2'], colors=['blue', 'orange'], alpha=0.6)
plt.xlabel('Zeit')
plt.ylabel('Werte')
plt.title('Gestapeltes Flächendiagramm')
plt.legend()
plt.grid(True)
plt.show()
```

**Stemplot**

```python
# Stemplot (Stiel- und Blattdiagramm)
x_values = np.arange(1, len(sample) + 1)
plt.figure(figsize=(8, 5))
markerline, stemlines, baseline = plt.stem(x_values, sample, linefmt='gray', markerfmt='bo', basefmt=" ")
plt.xlabel('Index')
plt.ylabel('Werte')
plt.title('Stemplot der Daten')
plt.grid(True)
plt.show()
```

**Stripplot**

```python
# Stripplot
plt.figure(figsize=(6, 5))
sns.stripplot(data=[sample, sample2], palette=['blue', 'orange'], jitter=True)
plt.xticks([0, 1], ['Sample 1', 'Sample 2'])
plt.xlabel('Datensatz')
plt.ylabel('Werte')
plt.title('Stripplot der Daten')
plt.grid(True)
plt.show()
```
